```python
# Qual a distribuição de idade dos participantes da pesquisa? (Histograma)
df.Idade.hist(bins = 60)
plt.xlabel("Idade")
plt.ylabel("Quantidade de pessoas")
plt.title("Distribuição de Idade")
plt.show()
```

```python
# Qual é a distribuição de sexo dos participantes da pesquisa?
df['Sexo'].value_counts(normalize=True) * 100

labels = 'Mulheres', 'Homens'
sizes = [53.94, 46.05]
explode = (0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 
plt.show()
```

```python
# Agrupando os dados
df[df["Assintomatico"] == "Sim"]["Idade"].value_counts()
```

```python
# Demonstrar o distribuição racial em casos de covid!
df["Raça/Cor"].value_counts(normalize=True) * 100 

labels = 'Parda', 'Branca', 'Amarela', 'Preta', 'Indigena', 'Ignorado'
sizes = [49.76, 28.19, 15.26, 3.69, 3.01, 0.06]
explode = (0, 0, 0, 0, 0.1, 0.2)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 
plt.show()
```

```python
# Demonstrar o distribuição racial em casos com evolução ao Obito!
df[df['Evolução do Caso'] == "Obito"]['Raça/Cor'].value_counts(normalize=True) * 100

labels = 'Parda', 'Branca', "Outros"
sizes = [79.12, 20.87, 0]
explode = (0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 
plt.show()
```
